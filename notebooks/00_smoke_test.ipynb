**Goal:** prove the Colab ↔ GitHub ↔ W&B loop works before any real ML work.

**What it does:**
1. `pip install` the three cloud SDKs we'll rely on all phase.
2. Read `data/test_profiles_10k.csv` straight from the GitHub raw URL (no local checkout needed).
3. Log `{"n_rows": len(df)}` to a W&B run in project `hangpost-matching`.

**Colab secrets required** (key icon on the left rail):
- `WANDB_API_KEY` — from https://wandb.ai/authorize
- `HF_TOKEN` — from https://huggingface.co/settings/tokens (read scope is enough here)

Once the W&B run appears in your dashboard, Step 2 is done.

```python
import os

try:
    from google.colab import userdata
    os.environ["WANDB_API_KEY"] = userdata.get("WANDB_API_KEY")
    os.environ["HF_TOKEN"] = userdata.get("HF_TOKEN")
except Exception as e:
    print("Not running in Colab, or secrets not set. Falling back to env vars.")

assert os.environ.get("WANDB_API_KEY"), "WANDB_API_KEY is missing"
assert os.environ.get("HF_TOKEN"), "HF_TOKEN is missing"
```

```python
import pandas as pd

CSV_URL = (
    "https://raw.githubusercontent.com/ai-bryguy101/hangpost-app/"
    "claude/ml-profile-matching-9gtCr/data/test_profiles_10k.csv"
)

df = pd.read_csv(CSV_URL)
print(f"Loaded {len(df):,} rows, {df.shape[1]} columns")
df.head()
```

```python
import wandb

run = wandb.init(
    project="hangpost-matching",
    name="00-smoke-test",
    job_type="smoke",
    config={"csv_url": CSV_URL},
)
wandb.log({"n_rows": len(df), "n_cols": df.shape[1]})
run.finish()
```
